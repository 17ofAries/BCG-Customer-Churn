```python
import pandas as pd
```

```python
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('clean_data_after_eda.csv')
df.head()
```

```python
df.info()
```

```python
# Convert date columns to datetime
df['date_activ'] = pd.to_datetime(df['date_activ'])
df['date_end'] = pd.to_datetime(df['date_end'])
```

```python
# Extract year, month, day
df['activation_year'] = df['date_activ'].dt.year
df['activation_month'] = df['date_activ'].dt.month
df['activation_day'] = df['date_activ'].dt.day
```

```python
# Calculate contract duration in days
df['contract_duration_days'] = (df['date_end'] - df['date_activ']).dt.days
```

```python
# Create price difference features
df['price_diff_6m_peak_off_peak'] = df['var_6m_price_peak'] - df['var_6m_price_off_peak']
df['price_diff_year_peak_off_peak'] = df['var_year_price_peak'] - df['var_year_price_off_peak']
```

```python
# Convert 'has_gas' to binary
df['has_gas'] = df['has_gas'].apply(lambda x: 1 if x == 't' else 0)
```

```python
df.head()
```

```python
df.drop(columns=['id', 'channel_sales', 'origin_up'], inplace=True)
```
