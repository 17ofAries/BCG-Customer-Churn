<h3><b>Data Dictionary</b></h3>

client_data.csv<br>
<ul><li>id = client company identifier</li>
<li>activity_new = category of the company’s activity </li>
<li>channel_sales = code of the sales channel </li>
<li>cons_12m = electricity consumption of the past 12 months </li>
<li>cons_gas_12m = gas consumption of the past 12 months </li>
<li>cons_last_month = electricity consumption of the last month </li>
<li>date_activ = date of activation of the contract </li>
<li>date_end = registered date of the end of the contract </li>
<li>date_modif_prod = date of the last modification of the product </li>
<li>date_renewal = date of the next contract renewal </li>
<li>forecast_cons_12m = forecasted electricity consumption for next 12 months </li>
<li>forecast_cons_year = forecasted electricity consumption for the next calendar year </li>
<li>forecast_discount_energy = forecasted value of current discount </li>
<li>forecast_meter_rent_12m = forecasted bill of meter rental for the next 2 months </li>
<li>forecast_price_energy_off_peak = forecasted energy price for 1st period (off peak) </li>
<li>forecast_price_energy_peak = forecasted energy price for 2nd period (peak) </li>
<li>forecast_price_pow_off_peak = forecasted power price for 1st period (off peak) </li>
<li>has_gas = indicated if client is also a gas client </li>
<li>imp_cons = current paid consumption </li>
<li>margin_gross_pow_ele = gross margin on power subscription </li>
<li>margin_net_pow_ele = net margin on power subscription </li>
<li>nb_prod_act = number of active products and services </li>
<li>net_margin = total net margin </li>
<li>num_years_antig = antiquity of the client (in number of years) </li>
<li>origin_up = code of the electricity campaign the customer first subscribed to </li>
<li>pow_max = subscribed power </li>
<li>churn = has the client churned over the next 3 months</ul><br>

price_data.csv 
<ul><li>id = client company identifier </li>
<li>price_date = reference date </li>
<li>price_off_peak_var = price of energy for the 1st period (off peak) </li>
<li>price_peak_var = price of energy for the 2nd period (peak) </li>
<li>price_mid_peak_var = price of energy for the 3rd period (mid peak) </li>
<li>price_off_peak_fix = price of power for the 1st period (off peak) </li>
<li>price_peak_fix = price of power for the 2nd period (peak) </li>
<li>price_mid_peak_fix = price of power for the 3rd period (mid peak)</ul><br>

Note: some fields are hashed text strings. This preserves the privacy of the original data but the 
commercial meaning is retained and so they may have predictive power

```python
client_df.drop(columns=['id','channel_sales','origin_up'], axis=1, inplace=True)
price_df.drop('id', axis=1, inplace=True)
```

```python
client_df.info()
```

```python
client_df_obj = client_df.select_dtypes(include='object')
client_df_obj.head()
```

```python
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'], format='ISO8601', dayfirst=True)
client_df['date_end'] = pd.to_datetime(client_df['date_end'], format='ISO8601', dayfirst=True)
client_df['date_modif_prod'] = pd.to_datetime(client_df['date_modif_prod'], format='ISO8601', dayfirst=True)
client_df['date_renewal'] = pd.to_datetime(client_df['date_renewal'], format='ISO8601', dayfirst=True)
client_df['has_gas'] = client_df['has_gas'].map({'t': 1, 'f':0})

price_df['price_date'] = pd.to_datetime(price_df['price_date'])
```

```python
client_df['date_stayed'] = (client_df['date_end'] - client_df['date_activ']).dt.days
```

```python
# Electricity consumption in the past 12 months
plt.figure(figsize=(8, 6))
sns.histplot(client_df['cons_12m'], kde=True)
plt.title('Distribution of Electricity Consumption (12 Months)', fontsize=14)
plt.xlabel('Electricity Consumption (kWh)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gas consumption in the past 12 months
plt.figure(figsize=(8, 6))
sns.histplot(client_df['cons_gas_12m'], kde=True, color='lightcoral')
plt.title('Distribution of Gas Consumption (12 Months)', fontsize=14)
plt.xlabel('Gas Consumption (cubic meters)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Net margin distribution
plt.figure(figsize=(8, 6))
sns.histplot(client_df['net_margin'], kde=True, color='green')
plt.title('Distribution of Net Margin', fontsize=14)
plt.xlabel('Net Margin')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of electricity consumption (12 months)
plt.figure(figsize=(15, 6))
sns.boxplot(x=client_df['cons_12m'], color='skyblue')
plt.title('Boxplot of Electricity Consumption (12 Months)')
plt.xlabel('Electricity Consumption (kWh)')
plt.show()
```

```python
# Boxplot of gas consumption (12 months)
plt.figure(figsize=(15, 6))
sns.boxplot(x=client_df['cons_gas_12m'], color='lightcoral')
plt.title('Boxplot of Gas Consumption (12 Months)')
plt.xlabel('Gas Consumption (cubic meters)')
plt.show()
```

```python
# Boxplot of net margin
plt.figure(figsize=(15, 6))
sns.boxplot(x=client_df['net_margin'], color='green')
plt.title('Boxplot of Net Margin', fontsize=14)
plt.xlabel('Net Margin')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(data=client_df.select_dtypes(include=np.number).corr(), annot=True, fmt='.2f')
plt.title('Correlation Matrix of Client Data', fontsize=16)
plt.show()
```

```python
# Pairplot for selected numerical features
plt.figure(figsize=(10, 8))
sns.pairplot(client_df[['cons_12m', 'net_margin', 'num_years_antig', 'pow_max']])
plt.suptitle('Pairplot of Selected Features', y=1.02, fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=client_df, x='churn', palette='muted')
plt.title('Churn Status Distribution', fontsize=14)
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Line plot of monthly electricity consumption (cons_last_month) over time
plt.figure(figsize=(15, 6))
client_df.sort_values('date_activ', inplace=True)

sns.lineplot(x='date_activ', y='cons_last_month', data=client_df, color='blue')
plt.title('Electricity Consumption Over Time', fontsize=14)
plt.xlabel('Date of Activation')
plt.ylabel('Electricity Consumption (kWh)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of electricity consumption vs net margin
plt.figure(figsize=(8, 6))
sns.scatterplot(x='cons_12m', y='net_margin', data=client_df, color='darkcyan')
plt.title('Electricity Consumption vs Net Margin', fontsize=14)
plt.xlabel('Electricity Consumption (kWh)')
plt.ylabel('Net Margin')
plt.show()
```

```python
# Scatter plot of maximum power vs net margin
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pow_max', y='net_margin', data=client_df, color='darkorange')
plt.title('Max Power vs Net Margin', fontsize=14)
plt.xlabel('Max Power (kW)')
plt.ylabel('Net Margin')
plt.show()
```

```python
# Violin plot of electricity consumption for churned vs non-churned clients
plt.figure(figsize=(8, 6))
sns.violinplot(x='churn', y='cons_12m', data=client_df, palette='muted')
plt.title('Electricity Consumption by Churn Status', fontsize=14)
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Electricity Consumption (kWh)')
plt.show()
```

```python
# Joint plot of electricity consumption vs forecast consumption
sns.jointplot(x='cons_12m', y='forecast_cons_12m', data=client_df, kind='reg', height=8, color='purple')
plt.suptitle('Electricity Consumption vs Forecast Consumption', y=1.02, fontsize=14)
plt.show()
```
